```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('marketing_campaign_data_lengkap.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding menggunakan scikit-learn
label_encoder = LabelEncoder()

df['Education_encode'] = label_encoder.fit_transform(df['Education'])
df['Marital_Status_encode'] = label_encoder.fit_transform(df['Marital_Status'])
df['klasifikasi_umur_encode'] = label_encoder.fit_transform(df['klasifikasi_umur'])
```

```python
df[['Education', 'Marital_Status', 'klasifikasi_umur', 'Education_encode', 'Marital_Status_encode', 'klasifikasi_umur_encode']].sample(5)
```

```python
df.drop(['Unnamed: 0.1', 'Unnamed: 0', 'ID', 'Year_Birth', 'Kidhome', 'Teenhome', 'Dt_Customer', 'MntCoke', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases'], axis=1, inplace=True)
```

```python
df.drop(['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Z_CostContact', 'Z_Revenue'], axis=1, inplace=True)
```

```python
df.drop(['Education', 'Marital_Status', 'klasifikasi_umur'], axis=1, inplace=True)
```

## STANDARISASI

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
standardized_data = scaler.fit_transform(df[['Income', 'Recency', 'NumWebVisitsMonth', 'Registration_Year', 'Umur_Saat_Registrasi', 'Conversion_Rate', 'Jumlah_Anak', 'Total_Pembelian', 'Total_Campaign', 'Total_Pengeluaran', 'Education_encode', 'Marital_Status_encode', 'klasifikasi_umur_encode']])
```

```python
df_standardized = pd.DataFrame(standardized_data, columns=['Income_St', 'Recency_St', 'NumWebVisitsMonth_St', 'Registration_Year_St', 'Umur_Saat_Registrasi_St', 'Conversion_Rate_St', 'Jumlah_Anak_St', 'Total_Pembelian_St', 'Total_Campaign_St', 'Total_Pengeluaran_St', 'Education_encode_St', 'Marital_Status_encode_St', 'klasifikasi_umur_encode_St'])
```
